```python
import pandas as pd
df=pd.read_csv('combined NO2 ,rainfall,MI.csv')
```

```python
df
```

```python
from scipy.stats import pearsonr
```

```python
mm=df['mm']
Median=df['Median']
MI=df['MangroveIndex']
```

```python
corr=pearsonr(Median,MI)
corr
```

```python
import seaborn as sns
sns.lmplot(x="Median", y="MangroveIndex", data=df);
```

```python
import numpy as np
from scipy import signal
correlation = signal.correlate(Median,MI, mode="same")
lags = signal.correlation_lags(Median.size, MI.size, mode="same")
lag = lags[np.argmax(correlation)]
import matplotlib.pyplot as plt

fig, (ax_orig, ax_noise, ax_corr) = plt.subplots(3, 1, figsize=(4.8, 4.8))
ax_orig.plot(Median)
ax_orig.set_title('NO2 median')
ax_noise.plot(MI)
ax_noise.set_title('MangroveIndex')
ax_noise.set_xlabel('Sample Number')
ax_corr.set_title('Lag')
ax_corr.plot(lags, correlation)
plt.tight_layout()
```
